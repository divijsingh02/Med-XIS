# Data Augmentation

There are 3 random states available , you can set random to 1 or 2 or 3 in the below cell

Each random state will increase/decrease brightness , crop and flip randomly

Instructions:

i)  folder of train and train_labels should be in the same folder with this codefile or you can randomly change the destination

ii)  Create 2 empty folders train_aug and label_aug where the augmented images will be saved

iii)  paste the augmented images in train and train_label folders and train the model


```python
random=3 # you can set 1 or 2 or 3
```

Required Libraries and functions

```python
import os
# !pip install pytorch==1.8.0 torchvision==0.9.0 torchaudio==0.8.0 cudatoolkit=11.1 -c pytorch -c conda-forge

# os.environ['CUDA_LAUNCH_BLOCKING'] = '1'
import numpy as np
import torch

import torch.nn as nn
import torch.nn.functional as F

import torch.utils.data as data
import torchvision
from torchvision import transforms
import torch.optim as optim
import matplotlib.pyplot as plt
import cv2 
def decrease_brightness(img, value=40):
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    h, s, v = cv2.split(hsv)

    lim = value
    v[v < lim] = 0
    v[v >= lim] -= value

    final_hsv = cv2.merge((h, s, v))
    img = cv2.cvtColor(final_hsv, cv2.COLOR_HSV2BGR)
    return img
def increase_brightness(img, value=50):
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    h, s, v = cv2.split(hsv)

    lim = 255 - value
    v[v > lim] = 255
    v[v <= lim] += value

    final_hsv = cv2.merge((h, s, v))
    img = cv2.cvtColor(final_hsv, cv2.COLOR_HSV2BGR)
    return img
```

Augmentation

```python

c = 0;
images = [];
filenam =[];
folder='train/'
for filename in os.listdir(folder): # List all the filenames in the folde
    # print('nwname ',filename)
    c = c + 1;
    print(c);
    print(filename)
    
    img = cv2.imread(os.path.join(folder, filename)); # Join the link of the folder and filename


    img = cv2.resize(img, (400,400))
    if random==1:
        
        img = img[0:400, 0:360]
        img = cv2.resize(img, (400,400))
    
        if c%3==0:
            img=cv2.flip(img, 1)
            img = increase_brightness(img, value=30)
        elif c%3==1:
            img=cv2.flip(img, 0)
            img = decrease_brightness(img, value=30)
        else:
            img=cv2.flip(img, -1)



        cv2.imwrite('train_aug/' +'A'+filename, img)
    elif random==2:
        
        img = img[0:400, 40:400]
        img = cv2.resize(img, (400,400))
    
        if c%3==1:
            img=cv2.flip(img, 1)
            img = increase_brightness(img, value=30)
        elif c%3==2:
            img=cv2.flip(img, 0)
            img = decrease_brightness(img, value=30)
        else:
            img=cv2.flip(img, -1)



        cv2.imwrite('train_aug/' +'B'+filename, img)
    elif random==3:
        
        img = img[40:400, 0:400]
        img = cv2.resize(img, (400,400))
    
        if c%3==2:
            img=cv2.flip(img, 1)
            img = increase_brightness(img, value=30)
        elif c%3==0:
            img=cv2.flip(img, 0)
            img = decrease_brightness(img, value=30)
        else:
            img=cv2.flip(img, -1)



        cv2.imwrite('train_aug/' +'C'+filename, img)
  

c=0
folder='train_labels/'
for filename in os.listdir(folder): # List all the filenames in the folde
    # print('nwname ',filename)
    c = c + 1;
    print(c);
    print(filename)
    img = cv2.imread(os.path.join(folder, filename)); # Join the link of the folder and filename
    img = cv2.resize(img, (400,400))
    
    if random==1:
        
        img = img[0:400, 0:360]
        img = cv2.resize(img, (400,400))
    
        if c%3==0:
            img=cv2.flip(img, 1)
        elif c%3==1:
            img=cv2.flip(img, 0)
        else:
            img=cv2.flip(img, -1)



        cv2.imwrite('label_aug/' +'A'+filename, img)
    elif random==2:
        
        img = img[0:400, 40:400]
        img = cv2.resize(img, (400,400))
    
        if c%3==1:
            img=cv2.flip(img, 1)
        elif c%3==2:
            img=cv2.flip(img, 0)
        else:
            img=cv2.flip(img, -1)



        cv2.imwrite('label_aug/' +'B'+filename, img)
    elif random==3:
        
        img = img[40:400, 0:400]
        img = cv2.resize(img, (400,400))
    
        if c%3==2:
            img=cv2.flip(img, 1)
        elif c%3==0:
            img=cv2.flip(img, 0)
        else:
            img=cv2.flip(img, -1)



        cv2.imwrite('label_aug/' +'C'+filename, img)
  
    


    
```
